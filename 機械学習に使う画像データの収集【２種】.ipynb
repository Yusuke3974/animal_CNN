# Flikr APIによる画像スクレイピング

FlikrAPIはFlikrのAPIキーを取得すると画像やその他のデータを取得できるようになります。<br>
Flikrにまずはメールアドレスとパスワードを登録します。<br>

FlikrHP:https://www.flickr.com/ <br>
Flikrapiインストール:https://stuvel.eu/software/flickrapi/ <br>
FlikrAPIの取得: https://www.flickr.com/services/api/ <br>

pythonファイルを作ってターミナルで実行して画像を取得していくので、<br>
ここではpythonファイル（download.py）のコードだけを表記していきます。<br>

```python
from flickrapi import FlickrAPI
from urllib.request import urlretrieve
from pprint import pprint
import os, time, sys
```

```python
# 自分のAPIキー情報を格納
Key = '**********************'

Secret = '**********************'
```

```python
# download.pyのコード
# 待ち時間を 1 秒間に設定（スパム対策）
wait_time = 1

# コマンドライン引数の １番目にanimarunameを設定
animalname = sys.argv[1]
# [例]python download.py animalname(dog,catなど)で画像取得ができるようにする。


# 画像を保存するディレクトリを指定
savedir = "./" + animalname

# FlickrAPIのインスタンスを設定
flickr = FlickrAPI(Key, Secret, format='parsed-json')
result = flickr.photos.search(
    # 検索キーワード
    text = animalname,
    # 取得するデータ件数（取得したい枚数以上に設定をおすすめ）
    per_page = 1500,
    # データの種類　「写真」に設定
    media = 'photos',
    # データの並び(関連順にソートをする)
    sort = 'relevance',
    # コンテンツ規制（アダルトなどを表示しない）
    safe_search = 1,
    # 取得したいオプション
    extras = 'url_q, licence'
)

# 結果
photos = result['photos']
pprint(photos)

# ダウンロードを実行
for i, photo in enumerate(photos['photo']):
    url_q = photo['url_q']
    # 設定したディレクトリにjpeg形式で保存
    filepath = savedir + '/' + photo['id'] + '.jpg'
    # ファイルが同じならパス
    if os.path.exists(filepath): continue
    urlretrieve(url_q,filepath)
    time.sleep(wait_time)
```

上記はpycharmやVScodeなどでターミナルから実行することになります。<br>
**python download.py cat<br>
python download.py bear**<br>
とターミナルに打ち込んでダウンロードしていきます。<br>

ただし、**関係のない画像データまで含まれてしまう**ことがあります。<br>
フォルダを実際に確認して、__テーマに関係のないものを削除していく__作業が必要です。<br>
